```python
import math
def civ(cgpa):
  cv=np.zeros((2))
  mx=np.mean(cgpa)
  sx=np.std(cgpa)
  lx=len(cgpa)
  cv[0]=mx-1.96*sx/math.sqrt(lx)
  cv[1]=mx+1.96*sx/math.sqrt(lx)
  return cv
```

```python
def zvalue(x1, x2):
    mx1 = np.mean(x1)
    mx2 = np.mean(x2)
    if mx1 == 0 and mx2 == 0:
        return 0, 1
    sx1 = np.std(x1)
    sx2 = np.std(x2)
    n1 = len(x1)
    n2 = len(x2)

    zv = (mx1 - mx2) / (math.sqrt(sx1**2 / n1 + sx2**2 / n2))
    zv = np.abs(zv)

    pv = stats.norm.sf(zv) * 2
    return zv, pv
```

```python
from statsmodels.stats.weightstats import ztest

def zvalue1(x1, x2):
    zv, pv = ztest(x1, x2)
    return zv, pv
```

```python
def zall(data):
    zv = np.zeros((20))
    pv = np.zeros((20))

    in1 = np.where(data[:, -1] == 0)
    in2 = np.where(data[:, -1] == 1)

    for j in range(20):
        x1 = data[in1[0], j]
        x2 = data[in2[0], j]
        zv[j], pv[j] = zvalue(x1, x2)

    return zv, pv
```

```python
for i in range(1, 33):
    fname = floc + str(i) + '.csv'
    data = np.genfromtxt(fname, delimiter=',')

    zv, pv = zall(data)

    in1 = np.where(pv <= 0.05)[0]
    print(f"College=={i} No of Features={len(in1)}")

    datan = data[:, in1]
    datan = np.concatenate((datan, data[:, -1].reshape(-1, 1)), axis=1)

    fname = floc + str(32 + i) + '.csv'
    np.savetxt(fname, datan, delimiter=',', fmt='%f')
```

```python
def rst(data):
    zv = np.zeros((20))
    pv = np.zeros((20))

    in1 = np.where(data[:, -1] == 0)
    in2 = np.where(data[:, -1] == 1)

    for j in range(20):
        x1 = data[in1[0], j]
        x2 = data[in2[0], j]
        zv[j], pv[j] = stats.ranksums(x1, x2)

    return zv, pv
```

```python
for i in range(1, 33):
    fname = floc + str(i) + '.csv'

    try:
        data = np.genfromtxt(fname, delimiter=',')
        zv, pv = rst(data)

        in1 = np.where(pv <= 0.05)[0]
        print(f"College=={i} No of Features={len(in1)}")

        datan = data[:, in1]  # Select significant features
        datan = np.concatenate((datan, data[:, -1].reshape(-1, 1)), axis=1)  # Add the label column

        fname = floc + str(32 * 2 + i) + '.csv'
        np.savetxt(fname, datan, delimiter=',', fmt='%f')

    except Exception as e:
        print(f"Error processing file {fname}: {e}")
```

```python
def ccra(data):
    r = np.corrcoef(data[:, :-1], rowvar=False)
    r = np.abs(r)
    imp = np.ones(np.shape(data)[1] - 1)
    for i in range(np.shape(data)[1] - 1):
        if imp[i] == 1:
            in1 = np.where(r[i, i+1:] >= 0.7)
            imp[in1[0] + (i + 1)] = 0
    return imp

```

```python
imp = ccra(data)
in1 = np.where(imp == 1)
newData = data[:, in1[0]]


r_new = np.corrcoef(newData, rowvar=False)
r_new=abs(r_new)

for i in range(len(in1[0])-1):
    print(np.max(r_new[i,i+1:]))
```

```python
for i in range(1,33):
  fname=floc+str(i)+'.csv'
  data=np.genfromtxt(fname,delimiter=',')
  pv=ccra(data)
  in1=np.where(pv==1)
  print("College==",i,"No of Features=",len(in1[0]))
  datan=data[:,in1[0]]
  datan=np.concatenate((datan,data[:,-1].reshape(-1,1)),axis=1)
  fname=floc+str(32*3+i)+'.csv'
  np.savetxt(fname,datan,delimiter=',',fmt='%f')
```
