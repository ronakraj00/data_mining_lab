```python
import math
def civ(cgpa):
  cv=np.zeros((2))
  mx=np.mean(cgpa)
  sx=np.std(cgpa)
  lx=len(cgpa)
  cv[0]=mx-1.96*sx/math.sqrt(lx)
  cv[1]=mx+1.96*sx/math.sqrt(lx)
  return cv
```

```python
def zvalue(x1, x2):
    mx1 = np.mean(x1)
    mx2 = np.mean(x2)
    if mx1 == 0 and mx2 == 0:
        return 0, 1
    sx1 = np.std(x1)
    sx2 = np.std(x2)
    n1 = len(x1)
    n2 = len(x2)

    zv = (mx1 - mx2) / (math.sqrt(sx1**2 / n1 + sx2**2 / n2))
    zv = np.abs(zv)

    pv = stats.norm.sf(zv) * 2
    return zv, pv
```

```python
def zall(data):
    zv = np.zeros((20))
    pv = np.zeros((20))

    in1 = np.where(data[:, -1] == 0)
    in2 = np.where(data[:, -1] == 1)

    for j in range(20):
        x1 = data[in1[0], j]
        x2 = data[in2[0], j]
        zv[j], pv[j] = zvalue(x1, x2)

    return zv, pv
```

```python
def ccra(data):
    r = np.corrcoef(data[:, :-1], rowvar=False)
    r = np.abs(r)
    imp = np.ones(np.shape(data)[1] - 1)
    for i in range(np.shape(data)[1] - 1):
        if imp[i] == 1:
            in1 = np.where(r[i, i+1:] >= 0.7)
            imp[in1[0] + (i + 1)] = 0
    return imp

```

```python
def crossover(chr):
    np.random.shuffle(chr)
    noc=np.shape(chr)[0]
    nchr=np.zeros((int(noc/2),20))
    for i in range(0,int(noc/2)):
        nchr[i,0:10]=chr[2*i,0:10]
        nchr[i,10:]=chr[2*i+1,10:]
    return nchr
```

```python
def mutation(chr):
    noc=np.shape(chr)[0]
    a=np.random.randint(0,20,[noc,2])
    for i in range(0,noc):
        chr[i,a[i,0]]=1-chr[i,a[i,0]]
        chr[i,a[i,1]]=1-chr[i,a[i,1]]
    return chr
```

```python
def kmeans_algo(f,noc):
    kmeans=KMeans(n_clusters=noc,random_state=0)
    kmeans.fit(f.reshape(-1,1))
    return kmeans.labels_
```

```python
def fuzzy_algo(f,noc):
    f1=np.zeros((np.shape(f)[0],2))
    f1[:,0]=f
    fd=skfda.FDataGrid(f1)
    c3=FuzzyCMeans(n_clusters=noc,random_state=0).fit(fd)
    return c3.membership_degree_
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
kf = KFold(5)
def acc_model(i,data,j):
  np.random.shuffle(data)
  data[:,0:-1]=normalizedata(data[:,0:-1])
  xdata=data[:0:-1]
  for k in range(0,np.shape(xdata)[1]):
    if len(np.unique(data[:,k]))>1:
      xdata[:,k]=kmeans_algo(xdata[:,k],2)
    in1=np.where(np.sum(xdata,axis=0)>0)
  ydata=data[:,-1]
  if j==1:
    xdata=xdata[:,in1[0]]
    in1=np.where(np.mean(xdata,axis=0)<1)
    xdata=xdata[:,in1[0]]
  if j==2:
    xdatan=np.zeros((np.shape(xdata)[0],np.shape(xdata)[1]*2))
    for k in range(0,np.shape(xdata)[1]):
      if len(np.unique(data[:,k]))>1:
        xdatan[:,2*k:2*(k+1)]=fuzzy_algo(xdata[:,k],2)
    xdata=xdatan
    in1=np.where(np.sum(xdata,axis=0)>0)
    xdata=xdata[:,in1[0]]
    in1=np.where(np.mean(xdata,axis=0)<1)
    xdata=xdata[:,in1[0]]
  pred=np.zeros((np.shape(xdata)[0],2))
  for tr_ind,ts_ind in kf.split(xdata):
    trdata=xdata[tr_ind,:]
    tsdata=xdata[ts_ind,:]
    tract=ydata[tr_ind]
    tsact=ydata[ts_ind]
    model[i].fit(trdata,tract)
    pred[ts_ind,0]=model[i].predict(tsdata)
    pred[ts_ind,1]=tsact
  return accuracy_score(pred[:,1],pred[:,0])




```
